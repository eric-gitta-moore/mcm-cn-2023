```python
import matplotlib.pyplot as plt
import numpy as np

# 创建示例数据
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# 创建 3D 表面图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# 设置坐标轴范围
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)

# 设置坐标轴边界
ax.set_box_aspect([1, 1, 1])

# 关闭坐标轴边界线
ax.axis('off')

plt.show()
```
