```python
import random


def get(length):
    def generate_decreasing_random_list(length, min_value, max_value):
        if length <= 0 or min_value >= max_value:
            return []

        random_list = [random.uniform(min_value, max_value)]

        while len(random_list) < length:
            new_value = random.uniform(min_value, random_list[-1])
            random_list.append(new_value)

        return random_list

    min_value = 0  # 随机数的最小值
    max_value = 100  # 随机数的最大值

    decreasing_list = generate_decreasing_random_list(length, min_value, max_value)
    print(decreasing_list)
    return decreasing_list
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建模拟的扰动数据集
# np.random.seed(2)
total_iterations = 20
x = np.arange(1, total_iterations + 1)
data = np.log(x) + np.random.randn(total_iterations) * 0.2
data = -data + 5

# 添加一些抖动
jitter = np.random.randn(total_iterations) * 0.05
data_with_jitter = (data + jitter)

# 创建总循环次数的数组
iterations = np.arange(1, total_iterations + 1)

# 绘制折线图
plt.figure(figsize=(10, 6))
for i in range(10):
    plt.plot(
        iterations,
        data_with_jitter * np.array(get(total_iterations)).astype(float),
        marker="o",
        linestyle="-",
        markersize=5,
        label=f"扰动数据集{i}",
    )
plt.xlabel("总循环次数")
plt.ylabel("校验误差值")
plt.legend()
plt.grid(True)
plt.tight_layout()

exportEmf('扰动')
# 显示图形
plt.show()
```
