```python
def find_intersection(point1, point2, point3, point4):
    x1, y1 = point1
    x2, y2 = point2
    x3, y3 = point3
    x4, y4 = point4
    
    # 计算直线1的斜率
    m1 = (y2 - y1) / (x2 - x1) if x1 != x2 else float('inf')
    
    # 计算直线2的斜率
    m2 = (y4 - y3) / (x4 - x3) if x3 != x4 else float('inf')
    
    # 如果两条直线平行，则没有交点
    if m1 == m2:
        return None
    
    # 计算交点的x坐标
    x_intersect = (y3 - y1 + m1 * x1 - m2 * x3) / (m1 - m2)
    
    # 计算交点的y坐标
    y_intersect = m1 * (x_intersect - x1) + y1
    
    return (x_intersect, y_intersect)

# 示例调用
point1 = (1, 2)
point2 = (3, 2)
point3 = (5, 6)
point4 = (7, 8)

intersection = find_intersection(point1, point2, point3, point4)
if intersection:
    print(f"两条直线的交点为：{intersection}")
else:
    print("两条直线平行，没有交点。")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建模拟的扰动数据集
np.random.seed(0)
total_iterations = 100
data = np.cumsum(np.random.randn(total_iterations))

# 添加一些抖动
jitter = np.random.randn(total_iterations) * 5
data_with_jitter = data + jitter

# 创建总循环次数的数组
iterations = np.arange(1, total_iterations + 1)

# 绘制折线图
plt.figure(figsize=(10, 6))
plt.plot(iterations, data_with_jitter, marker='o', linestyle='-', markersize=5, label='扰动数据集')
plt.xlabel('总循环次数')
plt.ylabel('数据值')
plt.title('随着总循环次数减少的扰动数据集')
plt.legend()
plt.grid(True)

# 显示图形
plt.show()
```
